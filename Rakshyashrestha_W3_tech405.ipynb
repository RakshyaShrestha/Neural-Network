```python
sns.countplot(x='chocolate', data=df)
plt.title('Chocolate Candy Distribution')
plt.show()
```

```python
sns.histplot(df['sugarpercent'], kde=True)
plt.xlabel('Sugar Percent')
plt.ylabel('Frequency')
plt.title('Sugar Percent Distribution')
plt.show()
```

```python
sns.catplot(x='chocolate', y='pricepercent', data=df, kind='box')
plt.xlabel('Chocolate')
plt.ylabel('Price Percent')
plt.title('Price Percent by Chocolate')
plt.show()
```

```python
sns.catplot(x='chocolate', y='pricepercent', data=df, kind='violin')
plt.xlabel('Chocolate')
plt.ylabel('Price Percent')
plt.title('Price Percent by Chocolate')
plt.show()
```

```python
df['chocolate'].fillna(df['chocolate'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df = df.drop('competitorname', axis=1)
```

```python
df['chocolate'] = df['chocolate'].astype(int)
df['winpercent'] = df['winpercent'].astype(int)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
